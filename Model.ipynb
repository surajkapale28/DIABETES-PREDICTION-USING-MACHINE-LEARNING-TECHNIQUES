```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(pd.read_csv('diabetes.csv'))
```

```python
df
```

```python
df=df.replace({'Glucose':0, 'BloodPressure':0,'SkinThickness':0,'Insulin':0,'BMI':0,'DiabetesPedigreeFunction':0},np.NaN)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Glucose'].fillna(df['Glucose'].mean(), inplace=True)
df['BloodPressure'].fillna(df['BloodPressure'].mean(), inplace=True)
df['SkinThickness'].fillna(df['SkinThickness'].mean(), inplace=True)
df['Insulin'].fillna(df['Insulin'].mean(), inplace=True)
df['BMI'].fillna(df['BMI'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
x=df.drop(['Outcome'],axis=1)
y=df.Outcome
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='linear')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
prediction=model.predict([[5,121,72,23,112,26.2,0.245,30]])
if(prediction[0]==0):
    print("You could not have Diabetes")
else:
    print("You could have Diabetes")
```
